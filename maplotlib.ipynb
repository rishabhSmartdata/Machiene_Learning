```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5,6,7,8,9]
y =[50,89,54,68,78,32,78,98,45]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'g+--')
```

```python
x =[1,2,3,4,5,6,7,8,9]
y =[50,89,54,68,78,32,78,98,45]
z =[54,78,98,23,48,52,88,48,35]
a =[60,79,84,78,48,92,78,98,45]

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y , label="max")
plt.plot(x,z,label="Min")
plt.plot(x,a,label="Avg")
plt.legend(loc="best")
```
